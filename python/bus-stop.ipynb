```python
from bokeh.io import show
from bokeh.plotting import gmap
from bokeh.models import GMapOptions, ColumnDataSource, Circle
from bokeh.io import output_notebook
```

```python
output_notebook()
bokeh_width, bokeh_height = 800,600


api_key=''

def plot(lat, lng, zoom=15, map_type='roadmap'):
    gmap_options = GMapOptions(lat=lat, lng=lng, map_type=map_type, zoom=zoom)
    p = gmap(api_key, gmap_options, title='Taichuang Bus', width=bokeh_width, height=bokeh_height)
    p.circle([lng], [lat], size=10, alpha=0.5, color='red')
    
    
    show(p)
    return p
```

```python
import requests

site = 'https://ptx.transportdata.tw/MOTC/v2/Bus/RealTimeByFrequency/City/Taichung/310?%24top=30&%24format=JSON'
response = requests.get(site,headers={'User-agent': 'Mozilla/5.0'})
```

```python
import json
data = json.loads(response.text)
```

```python
print(json.dumps(data[0]['BusPosition'],sort_keys=True, indent=4))
```

```python
lat, lon = data[0]['BusPosition']['PositionLat'], data[0]['BusPosition']['PositionLon']
```

```python
p = plot(lat, lon)
```

```python
stopAPi = 'https://ptx.transportdata.tw/MOTC/v2/Bus/DisplayStopOfRoute/City/Taichung/310?%24top=30&%24format=JSON'
response = requests.get(stopAPi,headers={'User-agent': 'Mozilla/5.0'})
import json
data = json.loads(response.text)
```

```python
stopLng=[]
topLat=[]
for stop in data[0]['Stops']:
    topLat.append(stop['StopPosition']['PositionLat'])
    stopLng.append(stop['StopPosition']['PositionLon'])
```

```python



source = ColumnDataSource(
    data=dict(
        lat=topLat,
        lng=stopLng,
    )
)

circle = Circle(x="lng", y="lat",size=10,fill_color="blue", fill_alpha=0.8, line_color=None)

p.add_glyph(source, circle)
show(p)
```
